```python
import sympy as sp
import numpy as np

# 创建符号变量
x = sp.symbols('x')

# 定义函数
f = (2-x**4)/(3*x)

# 求导
derivative_f = sp.diff(f, x)
print("SymPy 求导结果:", derivative_f)
sp.integrate(derivative_f,x)

# # 求解方程
# solution = sp.solve(f, x)
# print("SymPy 求解方程结果:", solution)
```

```python
import numpy as np

# 定义函数
def f(x):
    return x**2 + 3*x + 2

# 使用NumPy的gradient函数求导
x_values = np.linspace(-10, 10, 100)
derivative_f = np.gradient(f(x_values), x_values)
print("NumPy 求导结果:", derivative_f)

# 使用NumPy的roots函数求解方程
coefficients = [1, 3, 2]
roots = np.roots(coefficients)
print("NumPy 求解方程结果:", roots)
```

```python
from scipy.optimize import fsolve

# 定义方程
def equation(x):
    return np.exp(-x)-x

# 初始猜测值
initial_guess = 1.0

# 使用 fsolve 求解方程
root = fsolve(equation, initial_guess)

# 打印结果
print(f"The root of the equation is: {root[0]}")

x = root[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, x_range=(-10, 10), num_points=100, title="Function Plot"):
    """
    绘制任意函数的图形。

    Parameters:
    - func: 要绘制的函数，是一个可以接受一个数值或数组作为输入并返回相应输出的函数。
    - x_range: x 轴的范围，是一个包含两个元素的元组。
    - num_points: 在指定范围内生成的点的数量。
    - title: 图形的标题。
    """
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = func(x_values)

    plt.plot(x_values, y_values, label=str(func.__name__))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    # plt.ylim(-3,3)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    # plt.show()

# 示例：绘制函数 y = x^2 在范围 [-5, 5] 内的图形
def quadratic_function(x):
    return np.exp(-x)-x


def straight(x):
    return x

plot_function(quadratic_function, x_range=(-5,5), title="Quadratic Function Plot")
# plot_function(straight, x_range=(-2, 2), title="Quadratic Function Plot")
```
